```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
files = ['img1.jpg', 'img2.jpg', 'img3.jpg']
k_values = [1, 2, 10, 30, 100]

for filename in files:
    img = Image.open(filename)
    img_arr = np.array(img, dtype=float)
    
    height, width, _ = img_arr.shape
    original_bytes = img_arr.size * 4 

    R = img_arr[:, :, 0]
    G = img_arr[:, :, 1]
    B = img_arr[:, :, 2]
    U_r, S_r, V_r = np.linalg.svd(R, full_matrices=False)
    U_g, S_g, V_g = np.linalg.svd(G, full_matrices=False)
    U_b, S_b, V_b = np.linalg.svd(B, full_matrices=False)
    

    plt.figure(figsize=(20, 10))
    plt.subplot(1, len(k_values) + 1, 1)
    plt.imshow(img)
    plt.title(f'\n{int(original_bytes/1024)} KB')
    plt.axis('off')

    for i in range(len(k_values)):
        k = k_values[i]

        rec_r = np.dot(U_r[:, :k], np.dot(np.diag(S_r[:k]), V_r[:k, :]))
        rec_g = np.dot(U_g[:, :k], np.dot(np.diag(S_g[:k]), V_g[:k, :]))
        rec_b = np.dot(U_b[:, :k], np.dot(np.diag(S_b[:k]), V_b[:k, :]))
        
        rec_img = np.dstack((rec_r, rec_g, rec_b))
        
        rec_img = np.clip(rec_img, 0, 255).astype(np.uint8)
        

        compressed_bytes = 3 * (height * k + k + k * width) * 4

        plt.subplot(1, len(k_values) + 1, i + 2)
        plt.imshow(rec_img)
        plt.title(f'k={k}\n{int(compressed_bytes/1024)} KB')
        plt.axis('off')
        
    plt.show()
```
